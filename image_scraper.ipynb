```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import requests
import io
import os
from PIL import Image
import time
wd = webdriver.Chrome()

Google_Image = \
    'https://www.google.com/search?site=&tbm=isch&source=hp&biw=1873&bih=990&'

def scrape_images(initial_url, wd, delay, max_images):
    def scroll(wd):
        try:
            wd.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(delay)
            print("Scrolling...")
        except Exception as e:
            print(f"Scrolling failed: {e}")
        
    url = initial_url
    wd.get(url)
    image_urls = set()
    skips = 0
        
    while len(image_urls) + skips < max_images:
        scroll(wd)
        thumbnails = wd.find_elements(By.CLASS_NAME, "rg_i.Q4LuWd")

        # prevent adding the same thumbnail over and over again
        for img in thumbnails[len(image_urls) + skips: max_images]:
            try:
                img.click()
                time.sleep(delay)
            except:
                continue
                
            images = wd.find_elements(By.CSS_SELECTOR, ".sFlh5c.pT0Scc.iPVvYb")
            #print(images)
            for image in images:
                if image.get_attribute('src') in image_urls:
                    max_images += 1
                    skips += 1
                    break
                    
                if image.get_attribute('src') and 'http' in image.get_attribute('src'):
                    image_urls.add(image.get_attribute('src'))
                    #print(image_urls)
                        
    return image_urls
                    
def download_image(download_path, url, file_name):
    try:
        image_content = requests.get(url).content
        image_file = io.BytesIO(image_content) # like storing a file in memory
        image = Image.open(image_file)  # Convert binary to an image
        file_path = download_path + file_name
        
        with open(file_path, "wb") as f:
            image.save(f, "JPEG")
            
        print("Success!")
    except Exception as e:
        print("Failed -", e)
```

```python
page = "https://www.google.com/search?sca_esv=1a8ff9d288a21f85&rlz=1C1ONGR_enUS1090US1090&sxsrf=ACQVn0-WXlyPmlRvH1hYKvFuCCTle5k0Vg:1710129011841&q=images+of+cats&tbm=isch&source=lnms&sa=X&ved=2ahUKEwj5gZWCp-uEAxVjIEQIHbswASYQ0pQJegQIChAB&biw=1280&bih=665&dpr=1.5"
urls = scrape_images(page, wd, 2, 10)
for i, url in enumerate(urls):
    download_image("cats/", url, str(i) + ".jpg")
print("Complete!")
wd.quit()
```

```python
data = input('Enter your search keywords: ')
num_images = int(input('Enter the number of images you want of each: '))
file_name = input('Enter file location: ')

data_list = data.split(", ")
count = 0
for i in data_list:
    search_url = Google_Image + 'q=' + i #'q=' because its a query
    print(search_url)
    urls = scrape_images(search_url, wd, 1, num_images)
    print(f"Found {len(urls)}")

    for j, url in enumerate(urls):
        download_image(file_name + "/", url, str(count) + ".jpg")
        count += 1
        
print("Complete!")
wd.quit()
```
